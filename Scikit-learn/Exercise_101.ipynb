Imagine we are interested in predicting penguins species based on their body measurement: culmen length and culmen depth. First we want to do some data exploration to get a feel for the data.  
Let's start by importing the required modules and load the data.

```python
# Standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Disable jedi autocompleter
%config Completer.use_jedi = False

sns.set(context='notebook', style='darkgrid')
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.info()
```

As we can see, there are 2 categorical and 4 numerical features. The `species` is our target variable. Since it is a categorical variable, it is a classification problem.   
Let's see how does our target variable looks like.

```python
penguins['species'].value_counts()
```

```python
num_cols = penguins.select_dtypes('number').columns
```

```python
penguins[num_cols].hist(figsize=(12, 8))
plt.tight_layout()
```

```python
sns.scatterplot(data=penguins, y='bill_length_mm', x='bill_depth_mm', hue='species')
```

```python
sns.pairplot(penguins, hue='species')
```
